# Importación de datos


```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



## En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
total_ingresos_tienda1 = tienda['Precio'].sum()
print(f"Ingreso total de la tienda 1: {total_ingresos_tienda1}")
```

```python
total_ingresos_tienda2 = tienda2['Precio'].sum()
print(f"Ingreso total de la tienda 2: {total_ingresos_tienda2}")
```

```python
total_ingresos_tienda3 = tienda3['Precio'].sum()
print(f"Ingreso total de la tienda 3: {total_ingresos_tienda3}")
```

```python
total_ingresos_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso total de la tienda 4: {total_ingresos_tienda4}")
```

# 2. Ventas por categoría

## En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
productos_vendidos_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Productos Vendidos por Categoría (Tienda 1):\n', productos_vendidos_por_categoria_tienda1.head())
```

```python
productos_vendidos_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Productos Vendidos por Categoría (Tienda 2):\n', productos_vendidos_por_categoria_tienda2.head())
```

```python
productos_vendidos_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Productos Vendidos por Categoría (Tienda 3):\n', productos_vendidos_por_categoria_tienda3.head())
```

```python
productos_vendidos_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Productos Vendidos por Categoría (Tienda 4):\n', productos_vendidos_por_categoria_tienda4.head())
```

# 3. Calificación promedio de la tienda


## En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio de la tienda 1: {calificacion_promedio_tienda1:.2f}")
```

```python
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio de la tienda 2: {calificacion_promedio_tienda2:.2f}")
```

```python
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio de la tienda 3: {calificacion_promedio_tienda3:.2f}")
```

```python
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio de la tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Gráfico de Barras: Ingreso Total por Tienda

ingresos_totales = {
    'Tienda 1': total_ingresos_tienda1,
    'Tienda 2': total_ingresos_tienda2,
    'Tienda 3': total_ingresos_tienda3,
    'Tienda 4': total_ingresos_tienda4
}

df_ingresos = pd.DataFrame(ingresos_totales.items(), columns=['Tienda', 'Ingreso Total'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Ingreso Total', data=df_ingresos, palette='viridis')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# 2. Gráfico de Barras: Calificación Promedio por Tienda

calificaciones_promedio = {
    'Tienda 1': calificacion_promedio_tienda1,
    'Tienda 2': calificacion_promedio_tienda2,
    'Tienda 3': calificacion_promedio_tienda3,
    'Tienda 4': calificacion_promedio_tienda4
}

df_calificaciones = pd.DataFrame(calificaciones_promedio.items(), columns=['Tienda', 'Calificación Promedio'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Calificación Promedio', data=df_calificaciones, palette='plasma')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5) # Calificaciones suelen ser de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Gráfico de Barras Agrupado: Productos Vendidos por Categoría (Top 5 por Tienda)

# Combinar las series de productos vendidos por categoría en un solo DataFrame
def get_top_categories(series, store_name, top_n=5):
    df = series.head(top_n).reset_index()
    df.columns = ['Categoría del Producto', 'Cantidad Vendida']
    df['Tienda'] = store_name
    return df

df_cat1 = get_top_categories(productos_vendidos_por_categoria_tienda1, 'Tienda 1')
df_cat2 = get_top_categories(productos_vendidos_por_categoria_tienda2, 'Tienda 2')
df_cat3 = get_top_categories(productos_vendidos_por_categoria_tienda3, 'Tienda 3')
df_cat4 = get_top_categories(productos_vendidos_por_categoria_tienda4, 'Tienda 4')

df_todas_categorias = pd.concat([df_cat1, df_cat2, df_cat3, df_cat4])

plt.figure(figsize=(14, 8))
sns.barplot(x='Categoría del Producto', y='Cantidad Vendida', hue='Tienda', data=df_todas_categorias, palette='viridis')
plt.title('Top 5 Categorías de Productos Vendidos por Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
